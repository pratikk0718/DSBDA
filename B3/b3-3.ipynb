```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tip.csv')
```

```python
df.rename(columns={'size': 'peoples'}, inplace=True)
```

```python
print(df.isna().sum())
```

```python
col_var = df[['peoples', 'sex', 'smoker', 'day', 'time']]
for i in col_var:
    print(i, df[i].unique())
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.lineplot(data=df, x='total_bill', y='tip')
plt.show()
```

```python
sns.lineplot(data=df, x='peoples', y='tip')
plt.show()
```

```python
sns.lineplot(data=df, x='peoples', y='tip', hue='time')
plt.show()
```

```python
sns.boxplot(data=df, x='day', y='total_bill')
plt.show()
```

```python
sns.boxplot(data=df, x='peoples', y='total_bill', hue='day')
plt.show()
```

```python
sns.boxplot(data=df, x='day', y='total_bill', hue='sex')
plt.show()
```

```python
sns.histplot(data=df, x='total_bill', hue='time')
plt.show()
```

```python
sns.histplot(data=df, x='tip', hue='sex')
plt.show()
```

```python
sns.histplot(data=df, x='tip', hue='smoker')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='smoker')
plt.show()
```

```python
sns.violinplot(data=df, x='day', y='tip')
plt.show()
```

```python
sns.violinplot(data=df, x='time', y='tip', hue='day')
plt.show()
```

```python
sns.violinplot(data=df, x='peoples', y='total_bill')
plt.show()
```
