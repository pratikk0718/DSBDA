```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('heart.csv')
```

```python
df = df.drop_duplicates()
```

```python
print(df.isna().sum())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), cmap='YlGnBu', annot=True)
plt.show()
```

```python
sns.lineplot(data=df, x="age", y="cp", hue="output")
plt.show()
```

```python
sns.lineplot(data=df, x="age", y="trtbps", hue="output")
plt.show()
```

```python
temp_df = df[['age', 'cp', 'trtbps', 'chol', 'fbs', 'oldpeak', 'output']]
sns.pairplot(temp_df, hue="output")
plt.suptitle("Looking for Insights in Data", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(temp_df.columns, 1):
    plt.subplot(4,3,i)
    plt.title(f"Distribution of {col} Data")
    sns.histplot(df[col], kde=True)
plt.tight_layout()
plt.show()
```
